# 1. Calculate the number of orders per month in 2018

```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Connect to DB
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="8356817383",
    database="ecommerce1"
)

cur = conn.cursor()

# SQL query
query = """
SELECT 
    MONTH(order_purchase_timestamp) AS months, 
    COUNT(order_id) AS order_count
FROM 
    orders 
WHERE 
    YEAR(order_purchase_timestamp) = 2018
GROUP BY 
    months
"""

# Execute query
cur.execute(query)
data = cur.fetchall()

# Close DB connection
cur.close()
conn.close()

# Create DataFrame
df = pd.DataFrame(data, columns=["months", "order_count"])

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="months", y="order_count", data=df, color = "red")
plt.title("Monthly Order Count in 2018")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.xticks(ticks=range(0, 12), labels=[
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])
plt.tight_layout()
plt.show()
```

# 2. Find the average number of products per order, grouped by customer city.

```python
import mysql.connector
import pandas as pd

# 1. Connect to the database
conn = mysql.connector.connect(
    host="localhost",      # or your DB host
    user="root",  # replace with your MySQL username
    password="8356817383",  # replace with your MySQL password
    database="ecommerce1"   # replace with your DB name
)

# 2. Create a cursor
cur = conn.cursor()

# 3. Define the query
query = """ WITH count_per_order AS (
    SELECT 
        orders.order_id, 
        orders.customer_id, 
        COUNT(order_items.order_id) AS oc
    FROM orders 
    JOIN order_items ON orders.order_id = order_items.order_id
    GROUP BY orders.order_id, orders.customer_id
)

SELECT 
    customers.customer_city,
    ROUND(AVG(count_per_order.oc), 2) AS average_orders
FROM customers 
JOIN count_per_order ON customers.customer_id = count_per_order.customer_id
GROUP BY customers.customer_city;

"""

# 4. Execute the query
cur.execute(query)

# 5. Fetch results
data = cur.fetchall()

# 6. Display results
data


df = pd.DataFrame(data,columns = ["customer_city","average orders"])
df.head(20)
```

# 4. Calculate the retention rate of customers, defined as the percentage of customers who make another purchase within 6 months of their first purchase.

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Connect to the database
conn = mysql.connector.connect(
    host="localhost",      # or your DB host
    user="root",  # replace with your MySQL username
    password="8356817383",  # replace with your MySQL password
    database="ecommerce1"   # replace with your DB name
)

# 2. Create a cursor
cur = conn.cursor()

# 3. Define the query
query = """WITH a AS (
    SELECT 
        customers.customer_id,
        MIN(orders.order_purchase_timestamp) AS first_order 
    FROM 
        customers 
    JOIN 
        orders ON customers.customer_id = orders.customer_id
    GROUP BY 
        customers.customer_id
), 
b AS (
    SELECT 
        a.customer_id,
        COUNT(DISTINCT orders.order_id) AS next_order 
    FROM 
        a 
    JOIN 
        orders ON orders.customer_id = a.customer_id
    WHERE 
        orders.order_purchase_timestamp > a.first_order
        AND orders.order_purchase_timestamp < DATE_ADD(a.first_order, INTERVAL 6 MONTH)
    GROUP BY 
        a.customer_id
)

SELECT 
    ROUND(100 * COUNT(b.customer_id) / COUNT(a.customer_id), 2) AS repeat_purchase_rate_percentage
FROM 
    a 
LEFT JOIN 
    b ON a.customer_id = b.customer_id;

"""

# 4. Execute the query
cur.execute(query)

# 5. Fetch results
data = cur.fetchall()

# 6. Display results
data

```

# 5. Identify the top 3 customers who spent the most money in each year.

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Connect to the database
conn = mysql.connector.connect(
    host="localhost",      # or your DB host
    user="root",  # replace with your MySQL username
    password="8356817383",  # replace with your MySQL password
    database="ecommerce1"   # replace with your DB name
)

# 2. Create a cursor
cur = conn.cursor()

# 3. Define the query
query = """ select years,customer_id,payment,d_rank
from
(select year(orders.order_purchase_timestamp) years,
orders.customer_id,
sum(payments.payment_value) payment,
dense_rank() over(partition by year(orders.order_purchase_timestamp)
order by sum(payments.payment_value) desc) d_rank
from orders join payments
on payments.order_id = orders.order_id
group by year(orders.order_purchase_timestamp),
orders.customer_id) as a
where d_rank <= 3;
"""

# 4. Execute the query
cur.execute(query)

# 5. Fetch results
data = cur.fetchall()

# 6. Display results
data

df = pd.DataFrame(data, columns = ["years", "id","payment","rank"])
sns.barplot(x = "id", y = "payment" , data = df, hue = "years")
plt.xticks(rotation = 90)
plt.show()
```
